```python
#subplot

import numpy as np
a=np.array([44,55,66,77,88,99])
b=np.array([41,52,63,74,85,96])
c=np.array([14,15,16,17,18,19])
d=np.array([5,50,8,70,76,90])

plt.plot(a,d)

#plot1
plt.subplot(2,2,3)
plt.plot(a,d)

#plot2
plt.subplot(2,2,2)
plt.plot(c,d)



```
